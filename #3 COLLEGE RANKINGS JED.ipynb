```python
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv(r"C:\Users\gisse\OneDrive\Documents\College JE.csv")
b = pd.read_csv(r"C:\Users\gisse\Downloads\World University Rankings 2023 JED.csv")

merged_df = pd.merge(a, b, how='inner', on='Name of University')

merged_df = merged_df.sort_values(by='University Rank', ascending=True)

plt.figure(figsize=(12, 8))
plt.bar(merged_df['Name of University'], merged_df['University Rank'], color='skyblue')
plt.xlabel('University')
plt.ylabel('Ranking')
plt.title('University Ranking')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

merged_df.rename(columns={'Total Expend ': 'Total Expend'}, inplace=True)

merged_df = merged_df.sort_values(by='University Rank', ascending=True)

plt.figure(figsize=(12, 8))
plt.bar(merged_df['Name of University'], merged_df['Total Expend'], color='orange')
plt.xlabel('University')
plt.ylabel('Total Expenditure')
plt.title('Total Expenditure per University')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
print("The correlation coefficient of approximately -0.44 indicates a moderate negative correlation between University Rank and Total Expenditure. This suggests that, on average, as the ranking of a university increases (lower rank number), the Total Expenditure tends to decrease.")
```

```python
import matplotlib.pyplot as plt

merged_df.rename(columns={'Total Expend ': 'Total Expend'}, inplace=True)

merged_df = merged_df.sort_values(by='University Rank', ascending=True)

fig, ax1 = plt.subplots(figsize=(18, 8))  

ax1.bar(merged_df['Name of University'], merged_df['Total Expend'], color='orange', label='Total Expenditure')
ax1.set_xlabel('University')
ax1.set_ylabel('Total Expenditure', color='orange')
ax1.tick_params('y', colors='orange')

ax2 = ax1.twinx()

ax2.plot(merged_df['Name of University'], merged_df['University Rank'], color='skyblue', marker='o', label='University Rank')
ax2.set_ylabel('University Rank', color='skyblue')
ax2.tick_params('y', colors='skyblue')

plt.title('Total Expenditure and University Rank per University')
ax1.set_xticklabels(merged_df['Name of University'], rotation=45, ha='right') 

fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

plt.show()
```

```python
import pandas as pd

merged_df.rename(columns={'Total Expend ': 'Total Expend'}, inplace=True)

merged_df = merged_df.sort_values(by='University Rank', ascending=True)

print(merged_df[['Name of University', 'University Rank', 'Total Expend']])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(merged_df['Total Expend'], merged_df['University Rank'], color='blue', alpha=0.7)
plt.title('Total Expenditure vs. University Ranking')
plt.xlabel('Total Expenditure')
plt.ylabel('University Ranking')
plt.grid(True)
plt.show()
```

```python
correlation_coefficient = merged_df['Total Expend'].corr(merged_df['University Rank'])
print(f"Correlation Coefficient: {correlation_coefficient}")
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.regplot(x='Total Expend', y='University Rank', data=merged_df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Total Expenditure vs. University Ranking with Regression Line')
plt.xlabel('Total Expenditure')
plt.ylabel('University Ranking')
plt.grid(True)
plt.show()
```

```python
print("This plot states how on average, as the total expenditure increases, the university ranking tends to decrease. In other words, universities with higher rankings tend to have lower total expenditures and vice versa.")
```
