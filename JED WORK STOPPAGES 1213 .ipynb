```python
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv(r"C:\Users\gisse\Downloads\WORKSTOPPAGES 2022.csv")
b = pd.read_csv(r"C:\Users\gisse\OneDrive\Documents\UNEMPLOYMENT RATE 2022 JED.csv")

merged_data = pd.merge(a, b, on='Month')

fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:blue'
ax1.set_xlabel('Month')
ax1.set_ylabel('Number of Stoppages', color=color)
ax1.bar(merged_data['Month'], merged_data['Number of stoppages beginning in the month'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Unemployment Rate', color=color)
ax2.plot(merged_data['Month'], merged_data['Unemploymenr Rate'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Relation between Work Stoppages and Unemployment Rate per Month')
plt.xticks(rotation=45, ha='right')  
plt.show()
```

```python
print("The p-value 0.3979 is associated with a statistical test of the null hypothesis that there is no correlation between the two variables. ")
```

```python
print("The correlation coefficient is negative (-0.3819), indicating a negative correlation between work stoppages and the unemployment rate.")
```

```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

a = pd.read_csv(r"C:\Users\gisse\Downloads\WORKSTOPPAGES 2022.csv")
b = pd.read_csv(r"C:\Users\gisse\OneDrive\Documents\UNEMPLOYMENT RATE 2022 JED.csv")

merged_data = pd.merge(a, b, on='Month')

correlation_coefficient, p_value = pearsonr(merged_data['Number of stoppages in effect in the month'], merged_data['Unemploymenr Rate'])

print(f'Correlation Coefficient: {correlation_coefficient}')
print(f'P-value: {p_value}')

fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:blue'
ax1.set_xlabel('Month')
ax1.set_ylabel('Number of Stoppages in Effect', color=color)
ax1.bar(merged_data['Month'], merged_data['Number of stoppages in effect in the month'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Unemployment Rate', color=color)
ax2.plot(merged_data['Month'], merged_data['Unemploymenr Rate'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Relation between Number of Stoppages in Effect and Unemployment Rate per Month')
plt.xticks(rotation=45, ha='right')  
plt.show()
```

```python
print("The correlation coefficient of -0.2804 indicates a negative correlation between the number of stoppages in effect in the month and the Unemployment Rate")
```

```python
print("The p-value of 0.5424 which is very high, meaning that there is not enough evidence to reject the null hypothesis that there is no significant correlation")
```
